```python
import numpy as pd
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("E:/data analysis/Diwali Sales Data.csv", encoding='ISO-8859-1')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.columns
```

```python
df.shape
```

```python
df.drop_duplicates().sum()
```

```python
df.drop(columns=['Status','unnamed1'],inplace=True)
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt

plt.hist(df['Amount'], bins= 20,color='skyblue')
plt.title('distribution of amounts')
plt.xlabel('amount')
plt.ylabel('frequency')
plt.show()
```

this shows that how much amount is used by the people in purchases.

```python
gender_data = df.groupby('Gender')['Amount'].sum()
plt.bar(gender_data.index, gender_data.values, color=['skyblue', 'lightgreen'])
plt.title('Total Sales Amount by Gender')
plt.xlabel('Gender')
plt.ylabel('Total Amount')
plt.tight_layout()
plt.show()
```

this shows the amount spend by the male and female and which gender spend more amount.

```python
df.head()
```

```python
state_data = df.groupby('State')['Amount'].sum().sort_values(ascending=False)
plt.figure(figsize=(20,8))
plt.bar(state_data.index, state_data.values, color='orange')
plt.title('Top States by Total Sales')
plt.xlabel('Total Sales Amount')
plt.ylabel('State')
plt.tight_layout()
plt.show()
```

this show the total amount of the country spend.

```python
age_data = df.groupby('Age Group')['Amount'].sum().sort_values(ascending=False)

plt.figure(figsize=(8,5))
plt.bar(age_data.index, age_data.values, color=['orange','lightgreen'])
plt.title('Total Sales by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Amount')
plt.tight_layout()
plt.show()
```

this shows how much amount is spended by which age group.

```python
state_data= df.groupby('State')['Orders'].sum().sort_values(ascending=False)
plt.figure(figsize=(20,8))
plt.bar(state_data.index,state_data.values, color=['blue','lightgreen'])
plt.title('total orders gain by the country')
plt.xlabel('state')
plt.ylabel('orders')
plt.show()
```

this shows which country gains the higest order nd so on.

```python
plt.figure(figsize=(8, 6))
plt.scatter(df['Age'], df['Amount'], alpha=0.5, color=['blue'], edgecolors='black')
plt.title('Customer Age vs Purchase Amount')
plt.xlabel('Age')
plt.ylabel('Purchase Amount')
plt.grid(True)
plt.tight_layout()
plt.show()
```

this scatter plot shows us which age group has higest purchase amount.

```python
plt.figure(figsize=(20, 10))
plt.scatter(df['State'], df['Amount'], alpha=0.5, color=['blue'], edgecolors='black')
plt.title('Customer Age vs Purchase Amount')
plt.xlabel('states')
plt.ylabel('Purchase Amount')
plt.grid(True)
plt.tight_layout()
plt.show()
```

this shows which country has the higest amount of purchase.
